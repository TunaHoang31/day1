## 1. Ôn tập Numby
---
Thư viện Numpy là thư viện số, chuyên xử lý dữ liệu dạng mảng nhiều chiều

```python
#Gọi thư viện
import numpy as np

#tạo một ma trận các số nguyên ngâu nhiên trong khoảng từ 1-100
matran_A = np.random.randint(1,100, (10,15))
print(matran_A)
print ('kich thuoc cua ma tran:', matran_A.shape)
```

```python
# tách ma trận con số 1: lấy tất cả cac hàng và 10 cột đầu tiên
matran_A1 = matran_A[:,0:10]
matran_A1
```

```python
# tách ma trận con số 2: lấy tất cả cac hàng và 5 cột cuoi cung
matran_A2 = matran_A[:,10:]
matran_A2
```

```python
print ('kich thuoc cua ma tran:', matran_A1.shape)
print ('kich thuoc cua ma tran:', matran_A2.shape)
```

```python
#cách truy cập dl trong numpy
hangdau = matran_A[0,:]
hangdau
```

```python
hangcuoi = matran_A[-1,:]
hangcuoi
```

```python
phantu0 = matran_A[0,0]
phantu0
```

```python
phantucuoi = matran_A[-1,-1]
phantucuoi
```

## 2. Ôn tập matplotlib
---

Thư viện matplotlib là thư viện chuyên sử dụng để trực hóa - vẽ các biểu đồ

```python
#goi thu vien
import matplotlib.pyplot as plt

ten_khoa_x = ['K64', 'K65', 'K66', 'K67', 'K68']
so_hs_y = [450, 540, 459, 1233,2133]
#ve bieu do
```

```python
#ve bieu do dang cot thẳng đứng
plt.bar(ten_khoa_x,
       so_hs_y,
       color='brown')
plt.title(' BIEU DO THE HIEN SO LUONH SV THEO TUNG KHOA')
plt.grid(ls='--')
plt.xlabel('so luong')
plt.ylabel('khó')
plt.show()
```

```python
#ve bieu do dang cot nằm ngang
plt.barh(ten_khoa_x,
       so_hs_y,
       color='black')
plt.title(' BIEU DO THE HIEN SO LUONH SV THEO TUNG KHOA')
plt.grid(ls='--')
plt.xlabel('so luong')
plt.ylabel('khóa')
plt.show()
```

```python
#ve bieu do dang đường
plt.plot(ten_khoa_x,
       so_hs_y,
        'r--d', )
       
plt.title(' BIEU DO THE HIEN SO LUONH SV THEO TUNG KHOA')
plt.grid(ls='--')
plt.xlabel('so luong')
plt.ylabel('khóa')
           
plt.show()
           
```

```python
#ve bieu do tròn
plt.pie(so_hs_y,
       labels= ten_khoa_x,
      autopct='%.2f%%',
      wedgeprops= dict(width=0.7, edgecolor='w'))
plt.title(' BIEU DO THE HIEN SO LUONH SV THEO TUNG KHOA')

plt.show()
```

```python
#Hiển thị 2 biểu đồ và cùng 1 khung
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
#ve bieu do dang cot thẳng đứng
plt.bar(ten_khoa_x,
       so_hs_y,
       color='brown')
plt.grid(ls='--')
plt.xlabel('so luong')
plt.ylabel('khó')

plt.subplot(1,2,2)
#ve bieu do tròn
plt.pie(so_hs_y,
       labels= ten_khoa_x,
      autopct='%.1f%%',
      wedgeprops= dict(width=0.7, edgecolor='w'))

plt.show()
```
